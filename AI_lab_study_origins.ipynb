##DEPTH FIRST SEARCH 

```python
import networkx as nx
```

```python
def dfs(graph,curr_node,visited):
  if(len(visited)==len(graph)):
    return
  print(curr_node,end=' ')
  visited.add(curr_node)
  for neighbour in graph[curr_node]:
    if(neighbour not in visited):
      dfs(graph,neighbour,visited)
```

```python
from networkx import DiGraph,draw
graph = DiGraph()
graph.add_edges_from([(0,1),(0,2),(0,3),(2,4),(2,5)])
draw(graph,with_labels=True)
dfs(graph,0,set())
```

## N-Queens

```python
def print_solution(mat):
  #print('in op')
  for row in mat:
    print(str(row).replace(',', '').replace('\'', ''))
  print()

def is_safe_to_place_queen(row,column,mat):
  #print('is safe?')
  for i in range(row):
    if(mat[i][column]=='Q'):
      #print('not safe')
      return False
  (i,j) = (row,column)
  while(i>=0 and j>=0):
    if(mat[i][j]=='Q'):
      #print('not safe')
      return False
    i -=1
    j -= 1
  (i,j) = (row,column)
  while(i>=0 and j<len(mat)):
    if(mat[i][j]=='Q'):
      #print('not safe')
      return False
    i -=1
    j +=1
  #print('safe')
  return True

def n_queens(mat,r):
  # row = len(mat)
  if(r==len(mat)):
    print_solution(mat)
    return -1
  for i in range(len(mat)):
    if(is_safe_to_place_queen(row=r,column=i,mat=mat)):
      mat[r][i] = 'Q'
      n_queens(mat,r+1)
    mat[r][i]='-'

n = 7 
mat = [['-' for x in range(n)]for y in range(n)]
# mat =  [['–' for x in range(N)]for y in range(N)] 
print(len(mat))
n_queens(mat,0)
```
