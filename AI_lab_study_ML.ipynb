```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets._samples_generator import make_blobs
```

```python
x,y = make_blobs(n_samples=300,centers=4)
```

```python
wcss=[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(wcss)
```

```python
kmeans = KMeans(n_clusters=4,init="k-means++", n_init=10, max_iter=300,random_state=0)
pred_y = kmeans.fit_predict(x)
plt.scatter(x[:,0],x[:,1])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s=300,color='pink')
```

###IMPLEMENTING ARTIFICIAL NEURAL NETWORKS FOR AN 
APPLICATION USING PYTHON - REGRESSION 

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=1000)
# X.shape,y.shape=((1000,100),(1000,))
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
clf = MLPRegressor()
clf.fit(X_train,y_train)
print(clf.score(X_test,y_test))
```

###IMPLEMENTING ARTIFICIAL NEURAL NETWORKS FOR AN 
APPLICATION USING PYTHON - CLASSIFICATION

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
X,y = make_circles(n_samples=1000,noise=0.05)
X_train,X_test,y_train,y_test = train_test_split(X,y)

clf = MLPClassifier()
clf.fit(X_train,y_train)
print(clf.score(X_train,y_train))
y_predict = clf.predict(X_test)

fig,ax = plt.subplots(1,2)
sns.scatterplot(X_test[:,0],X_test[:,1],hue=y_predict,ax=ax[0])
sns.scatterplot(X_test[:,0],X_test[:,1],ax=ax[1])
```
